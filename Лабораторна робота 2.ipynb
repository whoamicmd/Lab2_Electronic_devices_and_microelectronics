З документації візьмемо вольт амперну характеристику на ділянці стабілізації (Діода BZX85B12). Перебудуємо цю характеристику у лінійний формат (рис. 2.1). 

```python
# Виконаємо графік за допомогою бібліотеки matplotlib.pyplot
import matplotlib.pyplot as plt
# Дані мін. сер. макс. для напруги та струму
I = [0.5, 20, 88]
U = [11.8, 12, 12.2]   
   
plt.figure(figsize=(6, 4))
plt.plot(U, I, color="black")

plt.title("Графік ВАХ")
plt.xlabel("Напруга, В")
plt.ylabel("Струм, мА")
plt.grid(True)
plt.savefig("images/plot_BAX.png")  # збереження графіка
plt.close()
```

<center><img src="images/plot_BAX.png" width="500"/><h4>Рисунок 2.1</h4></center>

За допомогою отриманого графіка знайдемо диференціальний опір для трьох струмів, мінімального, максимального та номінального, а також середню напругу стабілізації. 
$$
U_{cm} = U_{cm\ min} + \frac{U_{cm\ max} - U_{cm\ min}}{2} = 11.8 + \frac {12.2 - 11.8}{2} = 12\ В,
$$
$$
R_d = \frac{U_{cm\ \max} - U_{cm\ \min}}{I_{cm\ \max} - I_{cm\ \min}} = \frac{12.2 - 11.8}{88 - 0.5} = 4.81\ Ом.
$$

Проаналізуємо режим роботи стабілітрона в параметричному стабілізаторі напруги (рис. 2.2).

<center><img src="images/TINA_q6QrALirLS.png " width="450"/><h4>Рисунок 2.2</h4></center>

Розрахуємо струм через стабілітрон, враховуючи реальні параметри (Vz = 11 В, R1 = 680 Ω, Vin = 15 В, реальний виміряний струм 7.5 мА).
$$
V_{R1} = V_{in} - V_{z} = 15 - 11 = 4\ В,
$$
$$
I_{z} = \frac {V_{R1}}{R1} = \frac {4}{680} = 5.88\ мА,
$$
Реальний струм 7.5 мА вказує на скориговану Vz ≈ 9.9 В.
$$
V_{R1} = I_{z} \cdot R1 = 7.5 \cdot 10^{-3} \cdot 680 = 5.1\ В.
$$

Також розрахуємо потужність, що виділяється в резисторах R1, Rн та в стабілітроні VD1. 
$$
P_{R1} = I_{z} \cdot V_{R1} = 5.88 \cdot 10^{-3} \cdot 4 = 0.0235 \ Вт.
$$
$$
P_{R2} = \frac {V_{R2}^2}{R2} = \frac {11^2}{3600} = 0.0336 \ Вт.
$$
$$
P_{z} = I_{z} \cdot V_{z} = 5.88 \cdot 10^{-3} \cdot 11 = 0.0638 \ Вт.
$$

Розрахуємо діапазон змін вхідної напруги $ U_{in} $, при якому режим роботи стабілітрона перебуває в допустимих межах (струм через стабілітрон $ I_z $ між $ I_{z \min} = 2.2 \, \text{мА} $ та $ I_{z \max} = 100 \, \text{мА} $). Припустимо мінімальне навантаження ($ I_L \approx 0 $), номінальну напругу стабілізації $ V_z = 11 \, \text{В} $ (рис. 2.3).
$$
U_{in} = V_{z} + I_{z} \cdot R1 
$$

```python
# Імпортуємо бібліотеки
import numpy as np
import matplotlib.pyplot as plt

V_z = 11
I_z = np.arange(2.2, 100.2, 0.2) # Створюємо масив значень від 2.2 мА до 100 мА (включно) з кроком 0.2 мА.
U_in = np.zeros(len(I_z)) # Створюємо "пустий" масив для результатів, заповнений нулями.
R1 = 680

# Запускаємо цикл, який пройде по кожному значенню в масиві струмів I_z.
for i in range(len(I_z)):
    U_in[i] = V_z + (I_z[i] * 10 **-3) * R1 # Результат записуємо у відповідну комірку масиву U_in.
```

```python
plt.figure(figsize=(5, 4)) # Створює полотно для графіка розміром 5x4 дюйми.
plt.plot(U_in, I_z, color="black") # Малює графік.
plt.title("Діапазон змін вхідної напруги $U_{in}$") # Встановлює заголовок графіка.
plt.xlabel("Напруга, В") # Підписує горизонтальну (X) та вертикальну (Y) осі.
plt.ylabel("Струм, мА")
plt.xticks(np.arange(10, 90, step=5)) # Встановлює крок 5 для позначок на осі X (від 10 до 90).
plt.yticks(np.arange(0, 105, step=5)) # Встановлює крок 5 для позначок на осі Y (від 0 до 105).
plt.grid(True) # Вмикає відображення сітки.
plt.savefig("images/plot_U_in.png") # Зберігає готовий графік у файл 'plot_U_in.png' в папці 'images'.
plt.close() # Закриває графік, щоб звільнити пам'ять.
```

<center><img src="images/plot_U_in.png" width="500"/><h4>Рисунок 2.3</h4></center>

Знайдемо вихідну напруга стабілітрона 1N4105 при збільшенні температури з 25°C до 50°C ($\Delta T$ = 25°C). Температурний коефіцієнт напруги згідно з технічною документацією, для стабілітрона 1N4105 ТКН становить +0.08 %/°С або +8.8 мВ/°C. Розрахунок зміни напруги ($\Delta V_z$):
$$
\Delta V_z = ТКН \cdot V_z \cdot \Delta T = 0.0008 \cdot 11 \cdot 25 = 0.22\ В.
$$
Таким чином, нова напруга стабілізації при 50°C становитиме ≈ 11.22 В

Ємність C1 (10 µF) у схемі виконує роль фільтруючого конденсатора. Вона підключена паралельно стабілітрону Z1 (1N4105) і призначена для згладжування пульсацій і шумів вхідної напруги від джерела VS1 (15 В). Якщо видалити ємність C1, пульсації від VS1 можуть передаватися на вихід, (до ±0.1–0.5 В залежно від якості джерела).

#### Висновок
В результаті виконання лабораторної роботи досліджено принцип дії та параметри стабілітрона 1N4105 (аналог BZX85B12), побудована вольт-амперна характеристика з середньою напругою стабілізації 12 В та диференціальним опором Rd = 4.81 Ом. Розрахунки режиму параметричного стабілізатора показали струм через стабілітрон I_z ≈ 5.88 мА (теоретичний) з урахуванням реального значення 7.5 мА та скоригованої V_z ≈ 9.9 В, а також потужності елементів: P_R1 ≈ 0.0235 Вт, P_R2 ≈ 0.0336 Вт, P_z ≈ 0.0638 Вт. Визначено діапазон вхідної напруги U_in від ≈12.5 В до ≈79 В для струму I_z в межах 2.2–100 мА. Аналіз температурного впливу (ΔT = 25°C) підтвердив зростання V_z на 0.22 В через лавинний механізм пробою. Функція конденсатора C1 як фільтра пульсацій підкреслює її роль у стабільності схеми. Результати узгоджуються з технічною документацією та демонструють практичну застосовність стабілітронів для стабілізації напруги.
